```python
fg=px.histogram(data,x= 'Region',y=' Estimated Unemployment Rate (%)',color='Region',title='Unemployment Rate(State Wise) by Histogram', template='plotly')
fg.update_layout(xaxis={'categoryorder':'total descending'})
fg.show()
```

**2**. **Scatter Plot**

```python
colors=['orange','blue','Red']
Region=['Jammu & Kashmir','Andhra Pradesh','Bihar']

from re import X
for i in range(3):
  x= data[data['Region']==Region[i]]
  plt.scatter(x['longitude'], x['latitude'], c=colors[i], label=Region[i])
  plt.xlabel("longitude")
  plt.ylabel("latitude")
  plt.legend()
```

```python
colors=['red','orange','blue','green','yellow','black','pink']
Region=['Delhi','Maharashtra','Goa','Gujarat','Odisha','Sikkim','Tripura']

from re import X
for i in range(7):
  x= data[data['Region']==Region[i]]
  plt.scatter(x['Region.1'], x['latitude'], c=colors[i], label=Region[i])
  plt.xlabel("Region.1")
  plt.ylabel("latitude")
  plt.legend()
```

```python
colors=['red','orange','blue','green','yellow','black']
Region=['Delhi','Maharashtra','Goa','Gujarat','Odisha','Sikkim']

from re import X
for i in range(6):
  x= data[data['Region']==Region[i]]
  plt.scatter(x['longitude'], x['Region.1'], c=colors[i], label=Region[i])
  plt.xlabel("longitude")
  plt.ylabel("Region.1")
  plt.legend()
```
